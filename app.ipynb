```python
import pandas as pd
df = pd.read_csv('./output/_full_.csv').sort_values(by='fecha_utc')


import plotly.express as px
fig = px.density_mapbox(df, lat='lat', lon='lng', z='magnitud', radius=10,
                        center=dict(lat=-40, lon=-70), zoom=3,
                        mapbox_style="open-street-map",animation_frame='fecha_utc',
                        height=800
                       )
fig.update_layout(title_text='Visualizacion Sismos Diarios Chile', title_x=0.5)
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 1
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['easing']='quadratic'


fig.show()
```

```python
print(fig.layout.updatemenus[0].buttons[0].args[1]
)
```

## Renderiza HTML con Mapa

```python
import plotly.io as pio
pio.write_html(fig, file='index.html', auto_open=True)
```
